```python
# Find a factorial of a number using recursion
n=5

class solution:
    def fun(self,n):
        if n==0 or n==1:
            return 1
        return n*self.fun(n-1)

s=solution()
print(s.fun(4))
print(s.fun(5))
```

```python
# Reverse an array using the recursion

class solution:
    def fun(self,arr,l,r):
        if l>=r:
            return print(arr)
        arr[l],arr[r]=arr[r],arr[l]
        self.fun(arr,l+1,r-1)
    
    def revers(self,arr):
        l=0
        r=len(arr)-1
        self.fun(arr,l,r)

s=solution()
print(s.revers([1,3,4,5]))
```

```python
# Selection sort 
num=[4,5,9,2,1,8,0,10]

class solution:
# This is the only solution mentioned in Video
    def fun(self,num):
        n=len(num)
        for i in range(0,n):
            min=i
            for j in range(i+1,n):
                if num[j]<num[min]:
                    min=j
            num[min],num[i]=num[i],num[min]
        return num
    # TC=O(n**2)) SC=O(1)   Precise TC=(n(n+1))/2

s=solution()
print(s.fun(num))
```

```python
# Bubble Sort (reverse)
num=[4,5,9,2,1,8,0,10]
class solution:
    def fun(self,num):
        n=len(num)
        for i in range(n-2,-1,-1):
            is_swap=False
            for j in range(0,i):
                if num[j]>num[j+1]:
                    num[j+1],num[j]=num[j],num[j+1]
                    is_swap=True
            if is_swap ==False:
                return num
        return num
    # TC=O(n**2)   SC=O(1)   Precise TC=(n(n+1))/2
s=solution()

print(s.fun(num))
```

```python
# Insertion sort
nums=[4,5,9,2,1,8,0,10]
num=[0,1,2,3,4,5,6,7]
class solution:
    def fun(self,num):
        n=len(num)
        for i in range(1,n):
            key=num[i]
            j=i-1
            while j>=0 and num[j]>key:
                num[j+1]=num[j]
                j-=1
            num[j+1]=key
        return num
    # TC=O(n(n+1))/2  i.e TC=O(n**2)  SC=O(1)
    
s=solution()

print(s.fun(num))
```

```python
# Merge sort --Divide and conquer
arr=[4,5,9,2,1,8,0,10]
class solution:
    def merge_sort(self,arr):
        if len(arr)<=1:
            return arr
        mid=len(arr)//2
        left=arr[:mid]
        right=arr[mid:]
        left=self.merge_sort(left)
        right=self.merge_sort(right)
        return self.merge_array(left,right)
    
    def merge_array(self,left,right):
        i,j=0,0
        n,m=len(left),len(right)
        result=[]
        while i<n and j<m:
            if left[i]<=right[j]:
                result.append(left[i])
                i+=1
            else:
                result.append(right[j])
                j+=1

        while i<n:
            result.append(left[i])
            i+=1

        while j<m:
            result.append(right[j])
            j+=1

        return result
    

arr=[4,5,9,2,1,8,0,10]
s=solution()

print(s.merge_sort(arr))

# TC=O(n**2)  SC=O(n)

```

```python
# Best time to buy and sell the stock
price=[7,2,5,6,4,8,2,1,2]
class solution:

    # Simple solution
    def fun(self,price):
        n=len(price)
        max_profit=0
        value=0
        for i in range(0,n):
            for j in range(i+1,n):
                value=price[j]-price[i]
                if value>max_profit:
                    max_profit=value
        return max_profit
        # TC=O(n**2) SC=O(1)

    
    # Optimal solution 
    def func(self,price):
        n=len(price)
        minp=float("inf")
        max_profit=0
        for i in range(0,n):
            if price[i]<minp:
                minp=price[i]
            if max_profit<(price[i]-minp):
                max_profit=(price[i]-minp)
        return max_profit
    # TC=O(n)  SC=O(1)   

s=solution()

print(s.func(price))

```

```python
# Longest Consecutive sequence 
#  Given array is unsorted
nums=[1,99,101,98,2,5,3,99,100,1,1]

class solution:
    # Simple solution
    def fun(self,nums):
        n=len(nums)
        max=0
        for i in range(0,n):
            num=nums[i]
            count=1
            while num+1 in nums:
                count+=1
                num=num+1
            if count>max:
                max=count
                count=0
        return max
    # TC=O(n**2)  SC=O(n)


    # Optimal SOlution
    def Logest_Conse_Sequence(self,nums):
        n=len(nums)
        my_set=set()
        for i in range(0,n):
            my_set.add(nums[i])
        longest=0
        for num in my_set:
            if num-1 not in my_set:
                x=num
                count=1
                while x+1 in my_set:
                    count+=1
                    x+=1
                longest=max(longest,count)
        return longest
    # TC=O(n+n+n) i.e O(3n)  SC=O(n)
    
s=solution()

print(s.fun(nums))
print(s.Logest_Conse_Sequence(nums))
```

```python
# Rotate a matrix by 90 degree
matrix=[[5,0,3],[7,-10,9],[1,-52,6]]

# Brute force solution
class solution:
    def rotate_matrix(self,matrix):
        r=len(matrix)
        c=len(matrix[0])
        result=[[0]*c for _ in range(r)]
        count= c-1
        for i in range(0,r):
            for j in range(0,c):
                if count>=0:
                    result[j][count]=matrix[i][j]
            count-=1

        return result
    
    # TC=O(nm)   SC=O(nm)


    # Optimal solution

    # First do transform and then reverse each row
    def Op_rotate_Matrix(self,matrix):
        c=len(matrix[0])
        r=len(matrix)
        for i in range(0,r):
            for j in range(i+1,c):
                matrix[i][j],matrix[j][i]=matrix[j][i],matrix[i][j]

        for i in range(0,r):
            matrix[i].reverse()
        return matrix
    # TC=O(nm)+O(nm) i.eO(n**2)   SC==(1)


s=solution()
print(s.rotate_matrix(matrix))
print(s.Op_rotate_Matrix(matrix))
```
