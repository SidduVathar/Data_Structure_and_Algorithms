What is stack?
Can store any type of Data.

What mechanism is used?
LIFO [ Last in First Out]



```python
class stack:   # Apply stack mechanism at the end of array/list
    def __init__(self):
        self.item=[]  # Item is an array
    
    def is_empty(self):
        return len(self.item)==0
    
    def push(self,item):
        self.item.append(item)
        # TC=O(1)
    
    def pop(self):
        if len(self.item)==0:
            return " Cannot pop, stack is empty"
        x=self.item.pop()
        return x
    
    def top(self):
        if len(self.item)==0:
            return " Cannot top, stack is empty"
        return self.item[-1]
    
    def size(self):
        return len(self.item)
    

stack=stack()
stack.push(5)
stack.push(10)
stack.push(15)
print(f"Stack content={stack}")
print(f"Popped item={stack.pop()}")
print(f"Stack content={stack}")
print(f"Top item after pop={stack.top()}")
print(f"Stack is empty={stack.is_empty()}")
print(f"Size of the stack={stack.size()}")
```

Implimentation of Queues [FIFO] -First in First out

What is Queues?
Can store any type of Data.

What mechanism is used?
FIFO [ First in First Out]

Terms:
enqueue - The new value append
front - First elemnet of array need to out
rear - Last element of array
dequeu - The First out element
is_empty - Check the queue is empty or not



```python
class queue:   # Apply stack mechanism at the start of array/list
    def __init__(self):
        self.item=[]  # Item is an array
    
    def is_empty(self):
        return len(self.item)==0
    
    def enqueue(self,item):
        self.item.append(item)
    
    def dequeue(self):
        if len(self.item)==0:
            return " Cannot dequeue, queue is empty"
        x=self.item.pop(0)
        return x
    # TC=O(n)

    def front(self):
        if len(self.item)==0:
            return " Cannot front, queue is empty"
        return self.item[0]
    
    def rear(self):
        if len(self.item)==0:
            return " Cannot rear, queue is empty"
        return self.item[-1]
    
    def size(self):
        return len(self.item)
    

queue=queue()
queue.enqueue(5)
queue.enqueue(10)
queue.enqueue(15)
print(f"queue content={queue}")
print(f"dequeue item={queue.dequeue()}")
print(f"queue content={queue}")
print(f"front item after dequeue={queue.front()}")
print(f"rear item after dequeue={queue.rear()}")
print(f"Stack is empty={queue.is_empty()}")
print(f"Size of the queue={queue.size()}")
# print(f" print the stack={stack.Print()}")
```

```python
# Impliment Stack uing Queue :- Use Queue data Structure but works like a stack
from collections import deque

class stack_queue_implimentation:
    def __init__(self):
        self.queue=deque()

    def push(self,item):
        self.queue.append(item)
        for _ in range(len(self.queue)-1):
            self.queue.append(self.queue.popleft())

    def pop(self):
        if len(self.queue)==0:
            return "Stack is empty"
        return self.queue.popleft()
    
    def peek(self):
        if len(self.queue)==0:
            return " Stack is empty"
        return self.queue[0]
    
    def is_empty(self):
        return len(self.queue)==0
    
    def size(self):
        return len(self.queue)
    
s=stack_queue_implimentation()

s.push(100)
s.push(200)
s.push(300)
s.pop()
# s.peek()


```

```python
# Impliment Queue using the 2 Stack

class StackQueue:
    def __init__(self):
        self.st1=[]
        self.st2=[]

    def push(self,x):
        while self.st1:
            self.st2.append(self.st1.pop())
        self.st1.append(x)
        while self.st2:
            self.st1.append(self.st2.pop())
    # TC=O(2n)
        
    def pop(self):
        if not self.st1:
            print("Stack is empty")
            return -1
        top_element=self.st1.pop()
        return top_element
    
    def peek(self):
        if not self.st1:
            print("Stack is empty")
            return -1
        return self.st1[-1]
    # TC=O(1)

    def is_empty(self):
        return not self.st1
    
s=StackQueue()

s.push(1)
s.push(1)
s.push(2)
s.push(3)
s.push(4)

s.pop()
s.peek()
```

```python
# Impliment Satck using DLL
from collections import deque

class node:
    def __init__(self,value):
        self.value=value
        self.prev=None
        self.next=None

class Stack_DLL:
    def __init__(self):
        self.head=None
        self.tail=None

    def push(self,value):
        new_node=node(value)
        if self.head == None:
            self.head=new_node
        else:
            current=self.head
            while current.next is not None:
                current=current.next
            current.next=new_node
            self.tail=new_node
            new_node.prev=current
    
    def pop(self):
        if self.head==None:
            return "stack is empty"
        self.tail=self.tail.prev
        self.tail.next=None

    def top(self):
        return self.tail.value
    
    def size(self):
        current=self.head
        count=1
        while current.next is not None:
            count+=1
            current=current.next
        return count

s=Stack_DLL()

s.push(1)
s.push(2)
s.push(3)
s.push(4)
s.pop()
s.top()
s.size()
        
        

            

```
