```python
# Print all factors of a given number
# Simple solution
num=30
def fun(num):
    result=[]
    for i in range(1,num+1):
        if num%i==0:
            result.append(i)
    return result

fun(num)
# TC=O(n)

# Optimal solution
def func(num):
    result=[]
    for i in range(1,int(sqrt(num)+1)):
        if num%i==0:
            result.append(i)
            if (num//i)!=i:
                result.append(num//i)
    return result

func(num)
# TC=O(sqrt(n))
```

```python
# Find a factorial of a number
n=5

class solution:
    def fun(self,n):
        if n==0 or n==1:
            return 1
        return n*self.fun(n-1)

s=solution()
print(s.fun(4))
print(s.fun(5))
```

```python
# Reverse an array using the recursion

class solution:
    def fun(self,arr,l,r):
        if l>=r:
            return print(arr)
        arr[l],arr[r]=arr[r],arr[l]
        self.fun(arr,l+1,r-1)
    
    def revers(self,arr):
        l=0
        r=len(arr)-1
        self.fun(arr,l,r)

s=solution()
print(s.revers([1,3,4,5]))
```

```python
# Check given string is palindrome or not
class solution:
    def fun(self,s,l,r):
        if l>=r:
            print(True)
            return 
        if s[l]==s[r]:
            self.fun(s,l+1,r-1)
        else:
            print(False)
            return 
    def palin(self,s):
        l=0
        r=len(s)-1
        self.fun(s,l,r)

s=solution()
print(s.palin("sasasa"))
```

```python
# Selection sort 
num=[4,5,9,2,1,8,0,10]

class solution:
# This is the only solution mentioned in Video
    def fun(self,num):
        n=len(num)
        for i in range(0,n):
            min=i
            for j in range(i+1,n):
                if num[j]<num[min]:
                    min=j
            num[min],num[i]=num[i],num[min]
        return num
    # TC=O(n**2)) SC=O(1)   Precise TC=(n(n+1))/2

s=solution()
print(s.fun(num))
```

```python
# Bubble Sort (reverse)
num=[4,5,9,2,1,8,0,10]
class solution:
    def fun(self,num):
        n=len(num)
        for i in range(n-2,-1,-1):
            is_swap=False
            for j in range(0,i):
                if num[j]>num[j+1]:
                    num[j+1],num[j]=num[j],num[j+1]
                    is_swap=True
            if is_swap ==False:
                return num
        return num
    # TC=O(n**2)   SC=O(1)   Precise TC=(n(n+1))/2
s=solution()

print(s.fun(num))
```

```python
# Insertion sort
num=[4,5,9,2,1,8,0,10]
class solution:
    def fun(self,num):
        n=len(num)
        for i in range(1,n):
            key=num[i]
            j=i-1
            while j>=0 and num[j]>key:
                num[j+1]=num[j]
                j-=1
            num[j+1]=key
        return num
    # TC=O(n(n+1))/2  i.e TC=O(n**2)  SC=O(1)
    
s=solution()

print(s.fun(num))
```

```python
# Merge sort --Divide and conquer
num=[4,5,9,2,1,8,0,10]
class solution:
    def mergesort(self,num,l,r):
        if l<r:
            mid=(l+r)//2
            self.mergesort(num,l,mid)
            self.mergesort(num,mid+1,r)
            self.merge(num,l,mid,r)
    
    def merge(self,num,l,mid,r):
        left=num[l:mid+1]
        right=num[mid+1:r+1]
        i=0
        j=0
        k=l

        while i<len(left) and j<len(right):
            if left[i]<=right[j]:
                num[k]=left[i]
                i+=1
            else:
                num[k]=right[j]
                j+=1
            k+=1
        
        while i<len(left):
            num[k]=left[i]
            i+=1
            k+=1
        
        while j<len(right):
            num[k]=right[j]
            j+=1
            k+=1

s=solution()

s.mergesort(num, 0, len(num)-1)
print(num) # prints None



```

```python
# Two Sum Problem --Target is 8 return the index
arr=[5,9,1,2,3,4,15,6,3]

class solution:
    # Simple solutiom
    def fun(self,arr,k):
        n=len(arr)
        for i in range(0,n):
            for j in range(i+1,n):
                if i<=k and j<=k:
                    if arr[i]+arr[j]==k:
                        return [i,j]
        return -1
    # TC=O(n**2)  SC=O(1)


    # Optimal solution
    def func(self,arr,k):
        result={}
        n=len(arr)
        for i in range(0,n):
            if i<=13:
                remaining=k-arr[i]
                if remaining in result:  # The TC is O(1) for any value to check in dictionary
                    return [result[remaining],i]
                result[arr[i]]=i
    # TC =O(n)  SC=O(1)

s=solution()

print(s.func(arr,13))


```

```python
# Find the Maximum Subarray
arr=[-1000,-10,-100,-900,1,100,-99]

class solution:
    def fun(self,arr):
        n=len(arr)
        max=float("-inf")
        for i in range(0,n):
            count=arr[i]
            for j in range(i+1,n):
                count=count+arr[j]
            
            if max<count:
                max=count
        return max
    # TC=O(n**2)  SC=O(1)

    # Optimal solution
    def func(self,arr):
        n=len(arr)
        count=0
        max=float("-inf")
        for i in range(0,n):
            count+=arr[i]
            if count>max:
                max=count
            elif count<0:
                count=0
        return max

s=solution()

print(s.func(arr))
```
