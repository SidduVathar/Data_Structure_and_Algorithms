***Binary Search***
1. A real-life example where Bonary search works - Checking a word using the Dictionary book
2. Coding Problem
3. Iterative Solution
4. Recursive Solution
5. Time and Space Complexity

```python
# Binary search works when the given input is in sorted format  i.e other side of the Linear Search 

# Iterative Solution
arr=[2,4,6,7,7,9,11,18,19]

class solution:
    def Binary_Search(self,arr,target):
        n=len(arr)
        low=0
        high=n-1
        while low<=high:
            mid=(low+high)//2
            if arr[mid]==target:
                print(f"The target in index: {mid}")
                return
            elif arr[mid]<target:
                low=mid+1
            else:
                high=mid-1
        return -1
    
    # TC=O()  SC=O()

s=solution()

print(s.Binary_Search(arr,7))
```

```python
# Recursive solution

arr=[2,4,6,7,9,11,18,19]
class solution:
    def binary_search_recursive(self,arr,low,high,target):
        if low>high:
            # print(-1)
            return -1
        mid=(low+high)//2
        if arr[mid]==target:
            # print(mid)
            return mid
        elif arr[mid]<target:
            return self.binary_search_recursive(arr,mid+1,high,target)
        else:
            return self.binary_search_recursive(arr,low,mid-1,target)

s=solution()

print(s.binary_search_recursive(arr,0,len(arr)-1,2))
```

so, i.e log2(n)

    TC=O(log2(n))  i.e TC=O(5)  SC=O(1)

```python
# Implimenting lower bound [Smallest Index such that num[i]>=target]
# Upper bound [ num[i]>target ]

arr=[2,4,5,6,7,7,7,9,11,18,19]

class solution:
    def Binary_Search(self,arr,target):
        n=len(arr)
        low=0
        high=n-1
        lb=n
        while low<=high:
            mid=(low+high)//2
            if arr[mid]>=target:
                lb=mid
                high=mid-1
            else:
                low=mid+1
        return lb
    
    # TC=O(log2(n))  SC=O(1)

    def Upper_Bound_BS(self,arr,target):
        n=len(arr)
        low=0
        high=n-1
        ub=n
        while low<=high:
            mid=(low+high)//2
            if arr[mid]>target:
                ub=mid
                high=mid-1
            else:
                low=mid+1
        return ub
    
    # TC=O(log2(n))  SC=O(1)

s=solution()

print(s.Binary_Search(arr,3)) # Retruns the index
print(s.Upper_Bound_BS(arr,4))
print(s.Binary_Search(arr,7))
print(s.Upper_Bound_BS(arr,7))
```

```python
# CEIL THE FLOOR
# CEIL: Smallest no. in a array>=target / Floor:Largest no. in array<=target

arr=[2,4,5,6,7,7,7,9,11,18,19]
class solution:
    def ceil_floor(self,arr,target):
        n=len(arr)
        low=0
        high=n-1
        ceil=-1
        floor=-1
        while low<=high:
            mid=(low+high)//2
            if arr[mid]==target:
                return [arr[mid],arr[mid]]
            elif arr[mid]<target:
                floor=arr[mid]
                low=mid+1  
            else:
                ceil=arr[mid]
                high=mid-1


        return [floor,ceil]
        
    # TC=O(log2(n))  SC=O(1)

s=solution()

print(s.ceil_floor(arr,8))


```
