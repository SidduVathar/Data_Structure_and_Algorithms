```python
# Binary search works when the given input is in sorted format  i.e other side of the Linear Search 

# Iterative Solution
arr=[2,4,6,7,7,9,11,18,19]

class solution:
    def Binary_Search(self,arr,target):
        n=len(arr)
        low=0
        high=n-1
        while low<=high:
            mid=(low+high)//2
            if arr[mid]==target:
                print(f"The target in index: {mid}")
                return
            elif arr[mid]<target:
                low=mid+1
            else:
                high=mid-1
        return -1
    
    # TC=O()  SC=O()

s=solution()

print(s.Binary_Search(arr,7))
```

```python
# Recursive solution

arr=[2,4,6,7,9,11,18,19]
class solution:
    def binary_search_recursive(self,arr,low,high,target):
        if low>high:
            # print(-1)
            return -1
        mid=(low+high)//2
        if arr[mid]==target:
            # print(mid)
            return mid
        elif arr[mid]<target:
            return self.binary_search_recursive(arr,mid+1,high,target)
        else:
            return self.binary_search_recursive(arr,low,mid-1,target)

s=solution()

print(s.binary_search_recursive(arr,0,len(arr)-1,2))
```

```python
# Create a SLL CLASS
class node:
    def __init__(self,val):
        self.val=val
        self.next=None


class SLinkedList:
    def __init__(self) -> None:
        self.head=None

    def append(self,val):
        new_node=node(val)
        if self.head==None:
            self.head=new_node
        else:
            curr=self.head
            while curr.next is not None:
                curr=curr.next
            curr.next=new_node

    def Transverse(self):
        if self.head is None:
            print("SLL is empty")
        else:
            curr=self.head
            while curr is not None:
                print(curr.val,end=" ")
                curr=curr.next
            print()
    # TC=O(n)  SC=O(1)


# Insert at specific Position
    def insert_at(self,val,position):
        new_node=node(val)
        if position==0:
            new_node.next=self.head
            self.head=new_node  # So the new_node should be head of the SLL
        else:
            current=self.head
            prev_node=None
            count=0
            while current is not None and count <position:
                prev_node=current
                current=current.next
                count+=1
            prev_node.next=new_node
            new_node.next=current
     # TC=O(n)  SC=O(1)

# Delete the element given the index or the value
    def delete(self,val):
        temp=self.head
        if temp.next is not None:
            if temp.val==val:
                self.head=temp.next
                del temp
                return
            else:
                found=False
                prev=None
                while temp is not None:
                    if temp.val==val:
                        found=True
                        break
                    prev=temp
                    temp=temp.next
                
                if found:
                    prev.next=temp.next
                    del temp
                    return
                else:
                    print("Node not found")

        # TC=O(n)  SC=O(1)


n=SLinkedList()

# print(n.append(5))

n.Transverse()

n.append(5)
n.append(20)
n.append(12)
n.append(7)

n.insert_at(3,0)
n.insert_at(2,0)

n.Transverse()
        
```

```python
# Find the length of loop/cycle in a linked list
# The given data is head of the linked list

class node:
    def __init__(self,val):
        self.val=val
        self.next=None

class SLinkedList:
    def __init__(self) -> None:
        self.head=None

    
    def length_loop_Brute_Solution(self):
        my_dict=dict()
        temp=self.head
        travel=0
        while temp is not None:
            if temp in my_dict:
                return travel-my_dict[temp]
            my_dict[temp]=travel
            travel+=1
            temp=temp.next
        return 0
    # TC=O(n)   SC=O(n)


    def length_loop(self):
        slow=self.head
        fast=self.head
        while fast is not None and fast.next is not None:
            slow=slow.next
            fast=fast.next.next
            if fast==slow:
                slow=slow.next
                travel=1
                while slow!=fast:
                    slow=slow.next
                    travel+=1
                return travel                
        return 0
    # TC=O(n)  SC=O(1)
```

```python
# Find Pairs with Given Sum in Doubly Linked list
# Given DLL is Sorted and all elements are unique

class node:
    def __init__(self,val):
        self.val=val
        self.prev=None
        self.next=None

class doublyLinkedlist:
    def __init__(self):
        self.head=None

    # Better Solution
    def Sum_pair_DLL(self,target):
        temp=self.head
        my_set=set()
        result=[]
        while temp is not None:
                remaining=target-temp.val
                if remaining in my_set:
                    result.append([remaining,temp.val])
                my_set[temp.val]=remaining
                temp=temp.next

    # TC=O(n)  SC=O(n)

    # Optimal Solution
    def Sum_pair_dll(self,target):
        left=self.head
        result=[]
        right=self.head
        while right.next is not None:
            right=right.next
        while left.val<right.val:
            total=left.val+right.val
            if total==target:
                result.append([left.val,right.val])
                left=left.next
                right=right.prev
            elif total>target:
                right=right.prev
            else:
                left=left.next

        return result
    # TC=O(n)+O(n) i.e O(n)   SC=O(1)


             

```

```python
# Binary number conversion
# (9)_10 is decimal number -  To convert to binary number; Divide 9/2 First list the reminder then last the questiont i.e (1001)_275
# (13)_10 Binary number conversion: (1101)_2

def Convert2Binary(num:int)->str: # Given integer return in string format
    result=""
    while num>0:
        if num%2==1:
            result+="1"
        else:
            result+="0"
        num=num//2
    result=result[::-1]
    return result
    # TC=O(log_2(n))  SC=O(log_2(n))

print(Convert2Binary(13))
print(Convert2Binary(7))
```

```python
# From Binary to Decimal
# (1001)_2 -->(9)_10
# (1101)_2 -->(13)_10

def convert2Decimal(x:str)->int:
    decimal_number=0
    power=0
    index=len(x)-1
    while index>=0:
        num=int(x[index])*(2**power)
        decimal_number+=num 
        index-=1
        power+=1
    return decimal_number
    # TC=O(n)  SC=O(1)

convert2Decimal("0101")
```

```python
# Find the single number: Bit manipulation & XOR Trick

# Brute Solution
arr=[5,1,3,3,7,1,7]
def fun(arr):
    hasmap={}
    n=len(arr)
    for i in range(0,n):
        if arr[i] in hasmap:
            hasmap[arr[i]]=1
        else:
            hasmap[arr[i]]=0
    
    for key,val in hasmap.items():
        if val==0:
            print(key)
    # TC=O(n)  SC=O(n)

fun(arr)


# Optimal solution

def func(arr):
    ans=0
    for num in arr:
        ans=ans^num
    return ans

func(arr)
```

```python
# Power Set [Print all the subsets]

nums=[1,2,3]

def Powr_set(nums):
    n=len(nums)
    total_subset=1<<n
    result=[]
    for num in range(0,total_subset):
        lst=[]
        for i in range(0,n):
            if num&(1<<i)!=0:
                lst.append(nums[i])
        result.append(lst)
    return result

    # TC=O(n*2**n)
    # SC=O(n*2**n)
```

```python
# Subaray:  All the value must be continuous [1,2,3], [3,4,5,6]

# Subsequnce: a Continous/non-Continous sequence which follows the order.
# Example: [1,2,3,4] ,[2,3,4,7,8] , [1,2,4,5,10], [1,9] , [] empty can be subsequence
# Pick left to roght


# Print all the subsequences
nums=[9,5,7]

class Recursion:
    def Subsequence(self,ind:int,subset:list[int], nums:list[int], result:list[list[int]]):
        if ind>=len(nums):
            result.append(subset.copy())
            return
        subset.append(nums[ind])
        self.Subsequence(ind+1,subset,nums,result)
        subset.pop()
        self.Subsequence(ind+1,subset, nums, result)
   
    def subset(self,nums:list[int]) -> list[list[int]]:
        result=[]
        self.Subsequence(0,[],nums,result)
        return result
    # TC=O(2**n) n=len()
    # SC=O(n)

s=Recursion()

print(s.subset(nums))
```
